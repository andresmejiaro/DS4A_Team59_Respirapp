```python
def traduce_grupo(df):
    try:
        if df[['COD_DIAG']].isnull()['COD_DIAG']:
            df['diagnostico_traducido']=None
        else:
            code=icd10.find(df['COD_DIAG'])
            df['diagnostico_traducido']=code.block_description
    except:
        print(df['COD_DIAG'])
    return df
```
